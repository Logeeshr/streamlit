```python
def plot_sales(ax, df, item):
    plot_df = df.loc[df['item'] == item, 'date':'sales'].dropna()
    ax.plot(plot_df['date'], plot_df['sales'], label='Actual Sales')

    ax.set_xlabel('Date', size=8)
    ax.set_ylabel('Sales #', size=8)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, size=6)
    ax.set_yticklabels(ax.get_yticklabels(), size=6)
    ax.set_title(f'Sales for Item {item}', size=8)

# Create grid of subplots
fig, axs = plt.subplots(nrows=13, ncols=4, figsize=(25, 30))
axs = axs.flatten()

# Loop through each item number and plot sales data in each subplot
for i in range(1, 51):
    plot_sales(axs[i-1], df_agg, i)

# Add legend to last subplot
axs[-1].legend()

plt.tight_layout()
plt.show()
```
